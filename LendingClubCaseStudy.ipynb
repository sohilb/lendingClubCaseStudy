# 2. Data Cleaning and Manipulation

```python
# Check for missing values

missing_values = loandf.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

missing_values
```

```python
#dropping the columns with more than 40% null
drop_columns=loandf.columns[loandf.isnull().sum()/loandf.shape[0]*100 > 40]
drop_columns
```

```python
#Drop columns greater than 40% of null values
loandf.drop(columns = drop_columns, axis = 1, inplace=True)
loandf.shape
```

```python
# calculating the percentage of missing data in each column

percent_missing_values = (loandf.isnull().sum()/len(loandf))*100
percent_missing_values[percent_missing_values > 0]
```

```python
# Converting 'int_rate', 'revol_util' from object to float datatype
loandf['int_rate'] = loandf['int_rate'].str.strip('%').astype('float')
```

```python
# Converting 'revol_util' from object to float datatype
loandf['revol_util'] = loandf['revol_util'].astype(str).str.rstrip('%').astype(float) / 100 # first convert the datatype to string and then strip the '%'
loandf['revol_util'] = loandf['revol_util'].replace('', np.nan).astype(float) #replace empty string back to NaN
```

```python
# converting the 'issue_d', 'earliest_cr_line', 'last_pymnt_d ', 'last_credit_pull_d' to datetime format

loandf['issue_d'] = pd.to_datetime(loandf['issue_d'], format='%b-%y')
loandf['earliest_cr_line'] = pd.to_datetime(loandf['earliest_cr_line'], format='%b-%y')
loandf['last_pymnt_d'] = pd.to_datetime(loandf['last_pymnt_d'], format='%b-%y')
loandf['last_credit_pull_d'] = pd.to_datetime(loandf['last_credit_pull_d'], format='%b-%y')
```

### Fill emp_title missing values with mode

```python
# Fill missing value with mode for emp_title
loandf.emp_title.fillna(loandf.emp_title.mode()[0],inplace=True)
loandf.emp_title.info()
```

### Fill emp_legth with mode

```python
loandf.emp_length.fillna(loandf.emp_length.mode()[0], inplace=True)
loandf.emp_length.info()
```

```python
#analyse the fullypaid to defaulted percentages
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
total = len(loandf)
ax = sns.countplot(x="loan_status", hue="loan_status", data=loandf)
ax.set_xlabel('Loan Status')
ax.set_ylabel('Number of Loans')
plt.title('Distribution of Loan Status', fontsize=20)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.show()
```

# 4. Recommendations
Based on these findings, to mitigate the risk of loan defaults, the company could consider the following actions:

DTI Ratios: Applicants with high DTI ratios could be subjected to more stringent review processes.

Interest Rate Adjustments: Higher interest rates correlate with higher default rates. Adjusting lending criteria or pricing for these loans could help balance risk and return.

Annual Income (annual_inc): Lending club should do additonal judgements of those borrowers with relatively lower annual incone due to higher risk of default.

Loan Amount Limits: Implementing stricter limits or higher scrutiny for larger loan amounts could also help manage default risk.

Ideally the company should target borrowers with higher income but for relatively lower income borrowers loan default risk can be mitigated for them by providing lower loan amounts, have lower installments, lower interests etc
